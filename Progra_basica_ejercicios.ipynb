```python
# 26.- Comparar dos números
n1 = int(input("Dame el primer número: "))
n2 = int(input("Dame el segundo número: "))

if n1 > n2:
    print("El mayor es el primero:", n1)
elif n2 > n1:
    print("El mayor es el segundo:", n2)
else:
    print("Son exactamente iguales")
```

```python
# 27.- Cuadrado a la carta
opcion = input("¿Qué quieres calcular? Escribe 'area' o 'perimetro': ")
lado = float(input("¿Cuánto mide el lado del cuadrado? "))

if opcion == "area":
    print("El área es:", lado * lado)
elif opcion == "perimetro":
    print("El perímetro es:", lado * 4)
else:
    print("Error: Escribiste mal la opción (revisa mayúsculas/minúsculas)")
```

```python
# 28.- Mayor de edad
edad = int(input("¿Cuántos años tienes? "))

if edad >= 18:
    print("Pásale, eres mayor de edad.")
else:
    print("Alto ahí, eres menor de edad.")
```

```python
# 29.- División segura
dividendo = float(input("Número de arriba (dividendo): "))
divisor = float(input("Número de abajo (divisor): "))

if divisor == 0:
    print("¡Error! No se puede dividir entre cero.")
else:
    print("El resultado es:", dividendo / divisor)
```

```python
# 30.- Análisis de negocio
precio = float(input("Precio del producto: "))
cantidad = int(input("Cantidad vendida: "))
egresos = float(input("Total de gastos (egresos): "))

ingresos = precio * cantidad

if ingresos > egresos:
    print("¡Felicidades! Estás ganando dinero (Ganancias).")
elif ingresos == egresos:
    print("Estás tablas (Punto de equilibrio).")
else:
    print("Cuidado, estás perdiendo dinero (Pérdidas).")
```

```python
# 31.- Boleta de calificaciones
calificacion = float(input("Ingresa la calificación (0-10): "))

if calificacion < 0 or calificacion > 10:
    print("Error: Esa calificación no existe.")
elif calificacion < 6:
    print("Estatus: Irregular (Reprobado)")
elif calificacion < 10:
    print("Estatus: Regular")
else: # Si llegó aquí, es porque es 10 forzosamente
    print("Estatus: ¡Excelencia!")
```

```python
# 32.- ¿Par o Impar?
numero = int(input("Dime un número entero: "))

if numero % 2 == 0:
    print("El número es PAR")
else:
    print("El número es IMPAR")
```

```python
# 33.- Calculadora de Bonos
nombre = input("Nombre del vendedor: ")
ventas = float(input("¿Cuánto vendió en pesos? "))

if ventas < 1000:
    situacion = "Despedido (ponte las pilas)"
elif ventas < 5000:
    situacion = "En periodo de prueba"
elif ventas < 10000:
    situacion = "Te ganaste un Bono del 5%"
else:
    situacion = "¡Excelente! Bono del 10%"

print("Vendedor:", nombre)
print("Resultado:", situacion)
```

```python
# 34.- Clasificación de edad
edad = int(input("Ingresa tu edad: "))

if edad < 0 or edad > 120:
    print("Edad inválida (¿eres un vampiro o un viajero del tiempo?)")
elif edad < 10:
    print("Eres un Niño")
elif edad < 18:
    print("Eres un Adolescente")
elif edad < 30:
    print("Eres un Joven")
    print("Y también eres Mayor de Edad") # Agregamos esto como pide el ejercicio
elif edad < 60:
    print("Eres un Adulto")
    print("Y también eres Mayor de Edad")
else:
    print("Eres un Adulto Mayor")
    print("Y también eres Mayor de Edad")
```

```python
# 35.- Ordenar de mayor a menor
a = int(input("Número 1: "))
b = int(input("Número 2: "))
c = int(input("Número 3: ")
if a >= b and a >= c:
    mayor = a
    # Si a es el mayor, pelean b y c por el segundo puesto
    if b >= c:
        medio = b
        menor = c
    else:
        medio = c
        menor = b

elif b >= a and b >= c:
    mayor = b
    # Si b es el mayor, pelean a y c
    if a >= c:
        medio = a
        menor = c
    else:
        medio = c
        menor = a

else:
    mayor = c
    # Si c es el mayor, pelean a y b
    if a >= b:
        medio = a
        menor = b
    else:
        medio = b
        menor = a

print("Ordenados:", mayor, ",", medio, ",", menor)
```
