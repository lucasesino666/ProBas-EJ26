```python
# 37.- Interés compuesto con repetición (VERSIÓN CORREGIDA Y BLINDADA)
respuesta = "si"

# Usamos .lower() más abajo para que convierta todo a minúsculas automáticamente
while respuesta == "si" or respuesta == "sí":

    # Pedimos los datos
    c = float(input("Capital inicial (C): "))

    # Aclaramos cómo debe meter el dato para que no haya pierde
    i = float(input("Tasa de interés (Escribe el número normal, ej. 10 para 10%): "))
    n = int(input("Número de periodos (n): "))

    # TRUCO 1: Convertimos el porcentaje a decimal para que la fórmula matemática funcione
    tasa_decimal = i / 100

    # Fórmula correcta: M = C * (1 + i)**n
    m = c * (1 + tasa_decimal) ** n

    # Mostramos el resultado redondeado a 2 decimales para que parezca dinero real
    print("El monto final es: $", round(m, 2))

    # Preguntar al usuario si desea hacer otro cálculo
    respuesta = input("¿Deseas hacer otro cálculo? (si/no): ")

    # TRUCO 2: Convertimos su respuesta a minúsculas por si lo escribió con mayúsculas
    respuesta = respuesta.lower()

print("¡Programa terminado!")
```

```python
# 42.- Confirmar contraseña con 3 intentos máximo
contra1 = input("Escribe tu nueva contraseña: ")
contra2 = input("Confírmala (Intento 1 de 3): ")
intentos = 1

while contra1 != contra2 and intentos < 3:
    print("Las contraseñas no son iguales.")
    intentos = intentos + 1
    # Usamos str(intentos) para convertir el número a texto y poder imprimirlo
    contra2 = input("Confírmala (Intento " + str(intentos) + " de 3): ")

if contra1 == contra2:
    print("¡Contraseña guardada con éxito!")
else:
    print("Cuenta cancelada")
```

```python
# 43.- La tanda (guardar dinero hasta pasar de $100,000)
total_ahorrado = 0.0

while total_ahorrado <= 100000:
    abono = float(input("¿Cantidad a abonar? "))

    if abono < 0:
        print("Error: No se aceptan cantidades negativas.")
    else:
        total_ahorrado = total_ahorrado + abono
        print("Llevas ahorrado: $", total_ahorrado)

print("¡Felicidades, llegaste a la meta de los 100 mil!")
```

```python
# 44.- Calculadora que pregunta si quieres hacer otra cosa
respuesta = "si"

while respuesta == "si":
    n1 = float(input("Primer número: "))
    n2 = float(input("Segundo número: "))
    op = input("Elige la operación (+, -, *, /, **, %): ")

    if op == "+": print("Resultado:", n1 + n2)
    elif op == "-": print("Resultado:", n1 - n2)
    elif op == "*": print("Resultado:", n1 * n2)
    elif op == "/": print("Resultado:", n1 / n2)
    elif op == "**": print("Resultado:", n1 ** n2)
    elif op == "%": print("Resultado:", n1 % n2)
    else: print("Operación no existe.")

    respuesta = input("¿Deseas realizar otra operación? (Escribe 'si' o 'salir'): ")
```

```python
# 45.- Calculadora con dos ciclos. Uno repite la operación, otro repite el programa entero.
otra_operacion = "si"

while otra_operacion == "si":
    op = input("Elige la operación (+, -, *, /, **, %): ")
    repetir_misma = "si"

    # Este ciclo se repite solo si quieres seguir sumando (o restando, etc.)
    while repetir_misma == "si":
        n1 = float(input("Primer número: "))
        n2 = float(input("Segundo número: "))

        if op == "+": print("Resultado:", n1 + n2)
        elif op == "-": print("Resultado:", n1 - n2)
        elif op == "*": print("Resultado:", n1 * n2)
        elif op == "/": print("Resultado:", n1 / n2)
        elif op == "**": print("Resultado:", n1 ** n2)
        elif op == "%": print("Resultado:", n1 % n2)

        repetir_misma = input("¿Deseas repetir esta misma operación? (si/no): ")

    otra_operacion = input("¿Deseas realizar otra operación distinta? (si/no): ")
```

```python
# 48.- Llenar y buscar productos
n = int(input("¿Cuántos productos vas a guardar? "))
nombres = []
claves = []
existencias = []

# Primero guardamos los datos
for i in range(n):
    print("Datos del producto", i)
    nombres.append(input("Nombre: "))
    claves.append(input("Clave de barras: "))
    existencias.append(input("Cantidad de piezas: "))

# Luego buscamos por índice
indice = int(input("¿Qué índice deseas buscar? "))
print("--- DATOS DEL PRODUCTO ---")
print("Nombre:", nombres[indice])
print("Clave:", claves[indice])
print("Piezas en existencia:", existencias[indice])
```

```python
# 49.- Del anterior, pero guardando en un diccionario
nombres = ["Galletas", "Leche"]
claves = ["001", "002"]
existencias = [10, 5]

diccionario_productos = {} # Aquí está el diccionario vacío

for i in range(len(nombres)):
    # Usamos la clave como llave para abrir el candado del diccionario
    diccionario_productos[claves[i]] = {"Nombre": nombres[i], "Piezas": existencias[i]}

print("Mis productos guardados en el diccionario:")
print(diccionario_productos)
```

```python
# 50.- Buscar por nombre, clave o índice
nombres = ["Galletas", "Leche"]
claves = ["001", "002"]
existencias = [10, 5]

print("¿Cómo quieres buscar? (escribe: indice, nombre o clave)")
opcion = input("Opción: ")

if opcion == "indice":
    indice = int(input("Escribe el número del índice: "))
    print(nombres[indice], "| Clave:", claves[indice], "| Piezas:", existencias[indice])

elif opcion == "nombre":
    nombre_b = input("Escribe el nombre del producto: ")
    indice = nombres.index(nombre_b) # La computadora busca en qué posición está
    print(nombres[indice], "| Clave:", claves[indice], "| Piezas:", existencias[indice])

elif opcion == "clave":
    clave_b = input("Escribe la clave del producto: ")
    indice = claves.index(clave_b)
    print(nombres[indice], "| Clave:", claves[indice], "| Piezas:", existencias[indice])
```

```python
# 51.- Checador de asistencia
n = int(input("¿Cuántos trabajadores son? "))
nombres = []
asistencias = []

for i in range(n):
    nombres.append(input("Nombre del trabajador: "))
    asistencias.append(int(input("¿Vino? (1 para sí, 0 para no): ")))

for i in range(n):
    if asistencias[i] == 1:
        print(nombres[i], "asistió a trabajar.")
    else:
        print(nombres[i], "no asistió a trabajar.")
```

```python
# 53.- Ciclo while con una lista
numeros = []
respuesta = "si"

while respuesta != "no":
    num = float(input("Escribe un número: "))
    numeros.append(num)
    respuesta = input("¿Deseas seguir? (Escribe 'no' para terminar): ")

numeros.sort() # Esto los ordena de chiquito a grande
print("Lista ordenada:", numeros)
```

```python
# 55.- Menú maestro (simplificado y blindado)
tipo = input("¿Vas a guardar números o texto? (Escribe: numeros o texto): ")
lista = []
opcion = "0"

while opcion != "7":
    print("\n1. Agregar valores")
    print("2. Eliminar valores")
    print("3. Ordenar (modificando la original)")
    print("4. Ordenar (creando una nueva)")
    print("5. Verificar si existe un elemento")
    print("6. Calcular datos (sólo si elegiste numeros)")
    print("7. Salir")
    opcion = input("Elige una opción: ")

    if opcion == "1":
        dato = input("¿Qué guardamos? ")
        if tipo == "numeros": dato = float(dato)
        lista.append(dato)
        print("Tu lista:", lista)
    elif opcion == "2":
        dato = input("¿Qué borramos? ")
        if tipo == "numeros": dato = float(dato)
        if dato in lista: lista.remove(dato)
        print("Tu lista:", lista)
    elif opcion == "3":
        lista.sort()
        print("Ordenada:", lista)
    elif opcion == "4":
        nueva = sorted(lista)
        print("Nueva lista:", nueva)
    elif opcion == "5":
        dato = input("¿Qué buscas? ")
        if tipo == "numeros": dato = float(dato)
        if dato in lista: print("Sí está, en el índice:", lista.index(dato))
        else: print("No se encontró.")
    elif opcion == "6" and tipo == "numeros" and len(lista) > 0:
        print("Máximo:", max(lista), "Mínimo:", min(lista), "Suma:", sum(lista), "Promedio:", sum(lista)/len(lista))
    elif opcion == "7":
        print("¡Adiós!")
```
